```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from zipfile import ZipFile
from io import BytesIO
import gym
import gym_anytrading
from gym_anytrading.envs import StocksEnv
from stable_baselines3.common.vec_env import DummyVecEnv
from stable_baselines3 import PPO
```

```python
# Load data
df = pd.read_csv('BTCUSDT-1d-2025-05.csv', header=None)

# Rename columns for clarity
df.columns = ['Open Time', 'Open', 'High', 'Low', 'Close', 'Volume', 
              'Close Time', 'Quote Volume', 'Trades', 'Taker Buy Volume', 
              'Taker Buy Quote Volume', 'Ignore']

# Convert timestamps (microseconds to datetime)
df['Open Time'] = pd.to_datetime(df['Open Time'], unit='us', errors='coerce')
df['Close Time'] = pd.to_datetime(df['Close Time'], unit='us', errors='coerce')

# Drop invalid datetime rows
df = df.dropna(subset=['Open Time', 'Close Time']).reset_index(drop=True)

# Convert numeric columns
df[['Open', 'High', 'Low', 'Close', 'Volume']] = df[['Open', 'High', 'Low', 'Close', 'Volume']].astype(float)

# Visualizing Close Prices
plt.figure(figsize=(12, 6))
plt.plot(df['Close Time'], df['Close'], marker='o', linestyle='-', color='blue')
plt.title('BTC/USDT Close Prices (Daily - May 2025)')
plt.xlabel('Date')
plt.ylabel('Price (USDT)')
plt.grid(True)
plt.show()
```

```python
# Compute Simple Moving Average (3-period window)
df['SMA_3'] = df['Close'].rolling(window=3).mean()

# Compute Relative Strength Index (RSI)
def calculate_rsi(data, window=3):
    diff = data.diff(1).dropna()
    gain = diff.clip(lower=0)
    loss = -diff.clip(upper=0)
    avg_gain = gain.rolling(window=window).mean()
    avg_loss = loss.rolling(window=window).mean()
    rs = avg_gain / avg_loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

df['RSI'] = calculate_rsi(df['Close'])

# Drop rows with NaNs from indicators
df = df.dropna().reset_index(drop=True)

# Define custom trading environment
class CustomTradingEnv(StocksEnv):
    def _process_data(self):
        prices = self.df['Close'].to_numpy()
        signal_features = self.df[['Close', 'SMA_3', 'RSI']].to_numpy()
        return prices, signal_features
```

```python
# Set up environment for RL
window_size = 10
frame_bound = (window_size, len(df))
env = DummyVecEnv([lambda: CustomTradingEnv(df=df, frame_bound=frame_bound, window_size=window_size)])

# Train PPO model
model = PPO('MlpPolicy', env, verbose=1)
model.learn(total_timesteps=10000)

# Evaluate PPO model
obs = env.reset()
while True:
    action, _states = model.predict(obs)
    obs, rewards, done, info = env.step(action)
    if done:
        print("info:", info)
        break
```

```python
# Plot trading performance
plt.figure(figsize=(12, 6))
env.envs[0].render_all()
plt.show()
```
